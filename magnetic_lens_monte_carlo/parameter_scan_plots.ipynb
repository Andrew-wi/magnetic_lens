```python
import datetime
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# initialize variables
date = datetime.date.today()
# date = '2020-07-22'
l_cell_to_4k = 0.1
l_4k_to_beam_shutter = 0.26
mot_left_edge = 0.5500

# set seaborn plotting style
sns.set_style("darkgrid")

# import csv files and convert to pandas dataframe
scan_lens = pd.read_csv('manual_test_aug18_run1.csv')
scan_nolens = pd.read_csv('{}/no_lens_{}.csv'.format(date, date))

# create list of strings for column names
ints = list(range(0, 12))
ints = list(map(str, ints))

# add mean columns
scan_lens['mean'] = scan_lens[ints].mean(axis=1)
scan_nolens['mean'] = scan_nolens[ints].mean(axis=1)

# add std deviation columns
scan_lens['std_dev'] = scan_lens[ints].std(axis=1)
scan_nolens['std_dev'] = scan_nolens[ints].std(axis=1)

# add in distance from cell to 4k aperture
scan_lens['l_cell_to_lens_aperture'] = scan_lens['l_cell_to_lens_aperture'] + 0.1

# drop unwanted columns
scan_lens = scan_lens.dropna(axis='columns')

# display dataframe for error checking
scan_lens
```

```python
# pull relevant data from dataframes
dist_lens = scan_lens['l_cell_to_lens_aperture']
mean_lens = scan_lens['mean']
std_lens = scan_lens['std_dev']
mean_nolens = scan_nolens['mean'] / 500
std_nolens = scan_nolens['std_dev'] / 500

# plot results
plt.figure(figsize=(7*1.62, 7))
sns.lineplot(dist_lens, mean_lens, label='MSR, With Lens')
plt.fill_between(dist_lens, mean_lens + std_lens, mean_lens - std_lens, alpha=0.15)
plt.axhline(y=float(mean_nolens), ls=':', color='r', label='MSR, No Lens')

# add in beam shutter, MOT region distances
plt.axvline(x=l_cell_to_4k, color='c', ls='--', label='4K Aperture')
plt.axvline(x=mot_left_edge, color='g', ls='--', label='MOT Region')

# labels
plt.title('Mean Success Rate (MSR) vs. Lens Distance from Cell', fontsize=23)
plt.xlabel('Lens Position (m)', fontsize=15)
plt.ylabel('MSR', fontsize=15)
plt.tick_params(axis='both', labelsize=13)
plt.legend(fontsize=12)
plt.show()
```

```python
import datetime
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# set seaborn plotting style
sns.set_style("darkgrid")
```

```python
# initialize variables
date = datetime.date.today()
# date = '2020-07-22'
l_cell_to_4k = 0.1
l_4k_to_beam_shutter = 0.26
mot_left_edge = 0.5500
```

```python
# import csv files and convert to pandas dataframe
scan_lens = pd.read_csv('{}/lens_variation_{}.csv'.format(date, date))
scan_nolens = pd.read_csv('{}/no_lens_{}.csv'.format(date, date))

# create list of strings for column names
ints = list(range(0, 12))
ints = list(map(str, ints))

# add mean columns
scan_lens['mean'] = scan_lens[ints].mean(axis=1)
scan_nolens['mean'] = scan_nolens[ints].mean(axis=1)

# add std deviation columns
scan_lens['std_dev'] = scan_lens[ints].std(axis=1)
scan_nolens['std_dev'] = scan_nolens[ints].std(axis=1)

# display dataframe for error checking
scan_lens
```

```python
# pull relevant data from dataframes
dist_lens = scan_lens['l_cell_to_lens_aperture']
mean_lens = scan_lens['mean']
std_lens = scan_lens['std_dev']
mean_nolens = scan_nolens['mean']
std_nolens = scan_nolens['std_dev']

# plot results
plt.figure(figsize=(7*1.62, 7))
sns.lineplot(dist_lens, mean_lens, label='MSP, With Lens')
plt.fill_between(dist_lens, mean_lens + std_lens, mean_lens - std_lens, alpha=0.15)
plt.axhline(y=float(scan_nolens['mean']), ls='--', color='r', label='MSP, No Lens')

# add in beam shutter, MOT region distances
plt.axvline(x=l_cell_to_4k + l_4k_to_beam_shutter, color='k', ls=':')
plt.axvline(x=mot_left_edge, color='k', ls=':')

# labels
plt.title('Mean Successful Particle (MSP) vs Distance from Cell to Lens', fontsize=23)
plt.xlabel(r'$x$ (m)', fontsize=15)
plt.ylabel('MSP', fontsize=15)
plt.tick_params(axis='both', labelsize=13)
plt.legend()
plt.show()
```
