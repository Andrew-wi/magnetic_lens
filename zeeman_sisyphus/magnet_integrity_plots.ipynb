```python
# now along z; each z is 1 mm
x_pos = 15
y_pos = 15
plt_grad = range(0, 100)

x_comps_grad_along_z = [force_field[y_pos, x_pos, z][0] for z in plt_grad]
y_comps_grad_along_z = [force_field[y_pos, x_pos, z][1] for z in plt_grad]
z_comps_grad_along_z = [force_field[y_pos, x_pos, z][2] for z in plt_grad]

plt.figure(figsize = (12*1.62, 12))
plt.plot(plt_grad, x_comps_grad_along_z, label = 'x comp')
plt.plot(plt_grad, y_comps_grad_along_z, label = 'y comp')
plt.plot(plt_grad, z_comps_grad_along_z, label = 'z comp')
plt.ylabel('gradient of b-field (T/m)')
plt.xlabel('z-axis (mm)')
plt.legend()
plt.show()
```

```python
# norm of field along z
norm_along_z = [np.linalg.norm(force_field[y_pos, x_pos, z]) for z in points_to_plot]

plt.figure(figsize = (12*1.62, 12))
plt.plot(points_to_plot, norm_along_z, label = 'norm')
plt.legend()
plt.show()
```
